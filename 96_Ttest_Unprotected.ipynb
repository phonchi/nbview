```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
%matplotlib inline

trace1 = np.load("cr-rivfk_traces.npy")
trace2 = np.load("cr-rivrk_traces.npy")
test = np.load('test.npy')
```

```python
plt.plot(test[0])
```

```python
from scipy import stats
results = stats.ttest_ind(trace1, trace2, equal_var = False)
```

```python
resultarray = np.array(results)
resultarray.shape
```

```python
plotarray = resultarray[0,:]
```

```python
from matplotlib.backends.backend_pdf import PdfPages
with PdfPages('CR_Ttest.pdf') as pp:
    plt.plot(plotarray)
    U = np.empty(792)
    L = np.empty(792)
    U.fill(4.5)
    L.fill(-4.5)
    plt.plot(U)
    plt.plot(L)
    plt.ylim([-15, 15])
    pp.savefig(plt.gcf())
```
